## what brand sale the most Cruisers Bicycles

```python
%%sql
WITH Cruisers_b AS(
    SELECT * FROM products p
    JOIN categories c
    ON p.category_id = c.category_id
    WHERE category_name = "Cruisers Bicycles")

SELECT 
	b.brand_name, SUM(o.quantity)
FROM Cruisers_b cb
JOIN order_items o
ON cb.product_id = o.product_id
JOIN brands b
ON cb.brand_id = b.brand_id
GROUP BY 1
ORDER BY 2 DESC;
```

## what brand sale the most Mountain Bikes

```python
%%sql
WITH Mountain_b AS(
	SELECT * FROM products p
	JOIN categories c
 	ON p.category_id = c.category_id
	WHERE category_name = "Mountain Bikes")

SELECT 
	b.brand_name, SUM(o.quantity)
FROM Mountain_b mb
JOIN order_items o
ON mb.product_id = o.product_id
JOIN brands b
ON mb.brand_id = b.brand_id
GROUP BY 1
ORDER BY 2 DESC;
```

## total sale by category bike

```python
%%sql
SELECT 
    c.category_name,
    SUM(CASE 
			WHEN STRFTIME("%Y", o1.order_date) = '2016' THEN o2.quantity 
				ELSE 0 
		END
	) AS total_quantity_2016,
    SUM(CASE 
			WHEN STRFTIME("%Y", o1.order_date) = '2017' THEN o2.quantity 
				ELSE 0 
		END
	) AS total_quantity_2017,
    SUM(CASE 
			WHEN STRFTIME("%Y", o1.order_date) = '2018' THEN o2.quantity 
				ELSE 0 
		END
	) AS total_quantity_2018,
	SUM(o2.quantity) AS Total_quantity
FROM orders o1
JOIN order_items o2 ON o1.order_id = o2.order_id
JOIN products p
ON o2.product_id = p.product_id
JOIN categories c
ON p.category_id = c.category_id
GROUP BY c.category_name
ORDER BY Total_quantity DESC;

```

## what brand the most sale 

```python
%%sql
SELECT 
	b.brand_name , SUM(o.quantity)
FROM order_items o
JOIN products p
ON o.product_id = p.product_id
JOIN brands b
ON p.brand_id = b.brand_id
JOIN categories c
ON p.category_id = c.category_id
GROUP BY 1
ORDER BY 2 DESC;
```

## what product is top sales

```python
%%sql
SELECT 
	o.product_id, p.product_name, c.category_name, SUM(quantity) AS total_sale
FROM order_items o
JOIN products p
ON o.product_id = p.product_id
JOIN categories c
ON p.category_id = c.category_id
GROUP BY 1,2,3
ORDER BY 4 DESC
LIMIT 5;
```

```python
%%sql
SELECT 
	o.product_id, p.product_name, c.category_name, SUM(o.quantity) AS total_quantity,
    ROUND(SUM(o.list_price*o.quantity), 2) As total_sales, 
    ROUND(SUM((o.list_price*(1-o.discount)))*o.quantity, 2) AS total_sales_discount
FROM order_items o
JOIN products p
ON o.product_id = p.product_id
JOIN categories c
ON p.category_id = c.category_id
GROUP BY 1,2,3
ORDER BY 5 DESC
LIMIT 5;
```
